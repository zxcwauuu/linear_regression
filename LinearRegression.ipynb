# Импорты

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

# Данные

```python
data = pd.read_csv('datasets\cost_revenue_clean.csv')
data
```

# График данных

```python
fig, ax = plt.subplots(1, 1)
ax.set_xlabel('Бюджет')
ax.set_ylabel('Кассовый сбор')
ax.set_xlim(-0.1e8, 4e8)
ax.set_ylim(-0.1e9, 2.5e9)
ax.scatter(data['production_budget_usd'], data['worldwide_gross_usd'], alpha = 0.3)
```

# Модель

```python
model = LinearRegression()
```

```python
X = pd.DataFrame(data['production_budget_usd'])
Y = pd.DataFrame(data['worldwide_gross_usd'])
```

# Обучение модели

```python
model.fit(X, Y)
```

# Демонстрация линейной регрессии

```python
fig, ax = plt.subplots(1, 1)
ax.set_xlabel('Бюджет')
ax.set_ylabel('Кассовый сбор')
ax.set_xlim(-0.1e8, 4e8)
ax.set_ylim(-0.1e9, 2.5e9)
ax.scatter(data['production_budget_usd'], data['worldwide_gross_usd'], alpha = 0.3)
ax.plot(X, model.predict(X), color='red')
```

# Оценка R^2 модели

```python
model.score(X, Y)
```
